```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

```python
import numpy as np
import pandas as pd
```

```python
articles_df = pd.read_pickle('articles.pkl')
```

```python
articles_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

wordnet = WordNetLemmatizer()

def lemmatize(doc):
    return [wordnet.lemmatize(word) for word in word_tokenize(doc.lower())]

vectorizer = TfidfVectorizer(stop_words='english', max_features=2000, tokenizer=lemmatize)
vectors = vectorizer.fit_transform(articles_df.content)
words = vectorizer.get_feature_names()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans()

kmeans.fit(vectors)
```

```python
print('cluster centers:')
print(kmeans.cluster_centers_)
```

```python
top_centroids = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
print('top feature for cluster:')
for num, centroid in enumerate(top_centroids):
    print("%d: %s" % (num, ", ".join(words[i] for i in centroid)))
```
